```python
from typing import List
from pydantic import BaseModel, Field

class Outputformat(BaseModel):
  headline: str = Field(description="headline of the product")
  tagline: str = Field(description="tagline of the product")
  body_copy: str = Field(description="body copy of the product")
  CTA: str = Field(description="clear call to action")
class Responsevariation(BaseModel):
    variations: List[Outputformat] = Field(description="A list of 3 ad copy variations")

```

```python
import langchain
from langchain_openai import ChatOpenAI

llm = ChatOpenAI(
    model_name="gpt-5-nano",
    temperature =0.7,
    max_retries=2
).with_structured_output(Responsevariation)
```

```python
from langchain_core.prompts import ChatPromptTemplate

Prompt = ChatPromptTemplate.from_messages([
    (
        "system",
        """You are a product copywriter who has the knowledge of market trends and best go to catchy words for the product you'll be suggesting : headline, tagline, body copy, and CTA of the product
        Goal:deliver the best catchy headline, tagline, body copy, and CTA.
        """
    ),
    ("user", "{product_brief}")
])
```

```python
from langchain_core.prompts import ChatPromptTemplate
def generate_ad_copy(brief: str):
  chain = Prompt | llm
  try:
      response = chain.invoke({"brief:product_brief"})
      return response
  except Exception as e:
        return f"An error occurred: {e}"
```

```python
product_brief = 'New perfume for men, brand name: batman, target: 22-45 year old athletes'
result = generate_ad_copy(product_brief)

if isinstance(result, Responsevariation):
    print(result.model_dump_json(indent=2))
```
