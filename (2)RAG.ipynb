```python
from langchain_openai import OpenAI
from langchain_core.output_parsers import PydanticOutputParser
from langchain_core.pydantic_v1 import BaseModel, Field
import os

class ChatbotResponse(BaseModel):
    answer: str = Field(description="The answer to the question, based ONLY on the provided context. If the answer cannot be found in the context, state that explicitly.")
    source_document: str = Field(description="The name of the source document where the answer was found, or 'N/A' if the answer is not derived from the documents.")
    relevant_quote: str = Field(description="A relevant quote from the source document that supports the answer, or 'N/A' if the answer is not derived from the documents.")

def generate_with_source(query: str, context: str) -> ChatbotResponse:
    parser = PydanticOutputParser(pydantic_object=ChatbotResponse)
    format_instructions = parser.get_format_instructions()

    prompt_template = f"""
    You are a RAG chatbot designed to answer questions based ONLY on the provided context.
    Do NOT use any prior knowledge.

    If the answer to the question is directly and entirely found within the provided context,
    extract the answer, identify the source document name, and find a relevant quote from that source document that supports the answer.

    If the answer cannot be found in the provided context, you MUST respond by stating:
    "I cannot answer this question based on the provided documents."
    In this case, set 'source_document' and 'relevant_quote' to 'N/A'.

    Question: {query}

    Context:
    {context}

    {format_instructions}

    Your Response (in JSON format as specified by format_instructions):
    """

    llm = OpenAI(model="gpt-3.5-turbo-instruct", temperature=0) # Set temperature to 0 for more deterministic output
    response = llm.invoke(prompt_template)

    # Attempt to parse the response. Handle potential parsing errors gracefully.
    try:
        parsed_response = parser.parse(response)
    except Exception as e:
        # Fallback for parsing errors, or if LLM didn't follow format
        print(f"Warning: Could not parse LLM response. Error: {e}\nRaw response: {response}")
        # Create a default response indicating failure or inability to answer from context
        parsed_response = ChatbotResponse(
            answer="I cannot answer this question based on the provided documents due to an internal processing error.",
            source_document="N/A",
            relevant_quote="N/A"
        )
    return parsed_response
```

```python
from typing import Union

def jawab(query: str) -> str:

    if not store.get_all_documents():
        return "Error: Vector store is not initialized or contains no documents. Please ensure documents are loaded and embeddings are created successfully."

    results = store.similarity_search(query, k=3)

    context_parts = []
    for d in results:

        source_name = os.path.basename(d.metadata.get('source', 'Unknown Document'))
        context_parts.append(f"Content: {d.page_content}\nSource: {source_name}")

    context = "\n\n".join(context_parts)

    try:
        parsed_response = generate_with_source(query, context)

        formatted_output = f"Answer: {parsed_response.answer}\n" \
                           f"Source Document: {parsed_response.source_document}\n" \
                           f"Relevant Quote: \"{parsed_response.relevant_quote}\""
        return formatted_output
    except Exception as e:
        return f"An error occurred during response generation: {e}"
```

```python
query = "who is the prime minister of pakistan?"
answer = jawab(query)
print(answer)
```
